```python
import mysql.connector
import pandas as pd
import numpy as np

# Assuming df_1 is already defined in your notebook
# Replace NaN values with None
#df_1 = df_1.replace({np.nan: None})

# Ensure the DataFrame has the correct columns in the correct order
df_1 = df_1[['emp_id', 'name_1', 'name_2', 'occupation', 'salary', 'dept_id', 'bonus', 'total_salary', 'full_names', 'tax']]

# Convert DataFrame values to native Python types
df_1 = df_1.astype(object)

# Establish connection to MySQL
conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='Dennyd4116$'
)

# Create a cursor object
cursor = conn.cursor()

# Use the existing data warehouse parks_datawarehouse
cursor.execute('USE parks_datawarehouse')

# Drop the table if it already exists
cursor.execute('DROP TABLE IF EXISTS employee_salary')

# Create the table employee_salary
cursor.execute('''
CREATE TABLE employee_salary(
    emp_id INT,
    name_1 VARCHAR(50),
    name_2 VARCHAR(50),
    occupation VARCHAR(50),
    salary INT,
    dept_id INT,
    bonus INT,
    total_salary INT,
    full_names VARCHAR(100),
    tax INT
)
''')

# Insert the data into the table
for i in range(len(df_1)):
    cursor.execute('INSERT INTO employee_salary VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)', tuple(df_1.iloc[i]))

# Commit the changes
conn.commit()

# Make sure that the data has been inserted into the table
cursor.execute("SELECT * FROM employee_salary")
print(cursor.fetchall())

# Close the connection
conn.close()
```
